```python
# VISUALIZATION OF DATA 
```

```python
import matplotlib.pyplot as plot 
import seaborn as scn 
```

```python
df['Gold'].plot(kind='bar')
plot.xticks(rotation=90)
plot.xlabel=("Nations")
plot.ylabel=("Medals")
plot.title("GOLD MEDALS BY NATIONS")
plot.show()
```

```python
plot.figure(figsize=(10,6))
scn.histplot(df['Gold'],bins=30,kde=True)
plot.title("Gold medals")
plot.show()
```

```python
plot.figure(figsize=(14,8))
scn.barplot(x='NOC',y='Silver',data=df)
plot.xticks(rotation=90)
plot.tight_layout()
plot.xlabel = "NOC"
plot.ylabel = "Silver"
plot.title("Gold medals won by nations")
plot.tight_layout()
plot.show()
```
